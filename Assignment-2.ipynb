```python
#1. Reshape the array to 3 by 5
#2. Print array shape.
#3. Replace the max in each row by 0
import numpy as np
p = np.random.randint(1,21,15)
q=p.reshape(3,5)
print("reshaped array is ",q)
print("shape of array is ",q.shape)
maxNum = np.amax(q, axis=1)
np.where(np.isin(q,maxNum), 0, q)
```

```python
import pandas as pd
#Read "data.csv"
df=pd.read_csv("./data.csv")
#show the basic statistical description about the data
df.describe()
print("Basic statistical description about the data:")
print(df)
#check if the data has null values and replace the null values with the mean
df=df.fillna(df.mean())
print("Checking if the data has null values and replacing the null values with the mean:")
print(df)
#select atleast two columns  aggregate the data using: min, max, count, mean
print("I'm selecting all the columns & the aggregate is as below:")
print(df[['Duration', 'Pulse', 'Maxpulse', 'Calories']].agg(['min', 'max', 'mean', 'count']))
#filter the dataframe to select the rows with calories values between 500 and 1000
print("The rows with calories values between 500 and 1000:")
print(df[(df.Calories>500) & (df.Calories<1000)])
#filter the dataframe to select to select the rows with calories values > 500 and pulse < 100.
print("The rows with calories values > 500 and pulse < 100:")
print(df[(df.Calories>500) & (df.Pulse<100)])
#create a new "df_modified" dataframe that contains all the columns from df except for 'Maxpulse'
print("df_modified dataframe that contains all the columns from df except for Maxpulse:")
df_modified=pd.DataFrame(df,columns=["Duration", "Pulse", "Calories"])
print(df_modified)
#delete the 'maxpulse' column from the main df dataframe
del df['Maxpulse']
#convert the datatype of calories column to int datatype
df["Calories"] = df["Calories"].astype("int")
print("converting the datatype of calories column to int datatype:")
print(df)
#create a scatter plot for the two columns(Duration and calories).
df.plot.scatter(x= 'Duration', y= 'Calories')
print("scatter plot for the two columns, Duration and Calories:")
```

```python
import matplotlib.pyplot as plt
# Data to plot
Programminglanguages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
#I'm using different colors
colors = ["#DEB887", "#8A360F", "#5F9EA0", "#FF6103", "#7FFF00", "#FFF8DC"]
explode = (0.1, 0, 0, 0,0,0)  
# Plot
plt.pie(popuratity, explode=explode, labels=Programminglanguages, colors=colors, wedgeprops={'edgecolor' : 'black'},
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Popularity of programming languages')
plt.axis('equal')
plt.show()
```
